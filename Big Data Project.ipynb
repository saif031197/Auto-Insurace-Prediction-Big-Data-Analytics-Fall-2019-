```python
updated_data = og_data[og_data['Loss_Amount'] > 0]
```

```python
og_data = og_data[og_data['Loss_Amount'] == 0]
```

```python
og_data.shape
```

## Dividing Original Dataframe into chunks of 500

```python
max_rows = 478
m = 22
chunked_dataframes = []
while len(og_data) > max_rows:
    top = og_data[:max_rows]
    top = top.append(updated_data[:m])
    chunked_dataframes.append(top)
    og_data = og_data[max_rows:]
    updated_data = updated_data[m:]
else:
    top = (og_data)
    top = top.append(updated_data)
    chunked_dataframes.append(top)
```

## List of Dataframes

```python
len(chunked_dataframes)
```

```python
first_dataframe = chunked_dataframes[0]
```

```python
first_dataframe.shape
```

## Function to Compress Dataframe of 500 rows into one row and making a new dataframe of one row

```python
def compress_dataFrame(first):
    for row in first:
        first[row] = first[row].mean()
    first_new = first[:1]
    return first_new
```

## Appending Dataframes of one row each into a list of Dataframes

```python
merged_dataframes = []
for dataframe in chunked_dataframes:
    merged_dataframes.append(compress_dataFrame(dataframe))
```

```python
len(merged_dataframes)
```

## Merging the list of 849 dataframes into one final dataframe

```python
final_dataframe = pd.concat([df for df in merged_dataframes])
```
